```python
df.plot(kind = 'scatter', x = 'age', y = 'year');
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df, hue = 'age', height = 4) \
 .map(plt.scatter, 'age', 'nodes') \
.add_legend();
plt.show();

```

```python
plt.close()
sns.set_style('whitegrid');
sns.pairplot(df, hue = 'age', height = '4');
plt.show()
```

```python
sns.FacetGrid(df, hue="nodes", height=15) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```
